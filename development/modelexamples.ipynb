```python
x=numpy.linspace(-10,10,50)

y=numpy.linspace(-10,10,50)

xr, yr = numpy.meshgrid(x, y)

r=numpy.sqrt(xr**2 + yr**2)

z=numpy.sin(r)/(r*r)

zr= 10* (z/ (numpy.max(z)))

xm= xr + (zr/2)

ym= yr + (zr/2)

zeros=numpy.zeros(50)

fig= plt.figure()
ax = fig.add_subplot(1)

ax.plot_wireframe(xr, yr, zr,linewidth=1)

ax = fig.add_subplot(2)

ax.plot_wireframe(xm, ym,zeros, linewidth=1, "r")

ax.set_title('Drapped sheet model')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z') 
plt.show()
```
